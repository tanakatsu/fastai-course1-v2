```python
class SimpleNet(nn.Module):
    def __init__(self, layers):
        super().__init__()
        self.layers = nn.ModuleList([
            nn.Linear(layers[i], layers[i + 1]) for i in range(len(layers) - 1)])
        
    def forward(self, x): # (batch size, 3, 32, 32)
        x = x.view(x.size(0), -1) # (batch size, 3072)
        for l in self.layers:
            l_x = l(x)
            x = F.relu(l_x)
        return F.log_softmax(l_x, dim=-1)
```

## CNN

```python
class ConvNet(nn.Module):
    def __init__(self, layers, c):
        super().__init__()
        self.layers = nn.ModuleList([
            nn.Conv2d(layers[i], layers[i + 1], kernel_size=3, stride=2) # stride=2: same effect as 2 by 2 max-pooling
            for i in range(len(layers) - 1)])
        self.pool = nn.AdaptiveMaxPool2d(1) # output size is 1x1
        self.out = nn.Linear(layers[-1], c)
        
    def forward(self, x):
        for l in self.layers: x = F.relu(l(x))
        x = self.pool(x) #  (batchsize, 80, 3, 3) -> (batchsize, 80, 1, 1)
        x = x.view(x.size(0), -1)
        return F.log_softmax(self.out(x), dim=-1)
```

## Refactored

```python
class ConvLayer(nn.Module):
    def __init__(self, ni, nf):
        super().__init__()
        self.conv = nn.Conv2d(ni, nf, kernel_size=3, stride=2, padding=1) # added paddng=1
        
    def forward(self, x): return F.relu(self.conv(x))
```

```python
class ConvNet2(nn.Module):
    def __init__(self, layers, c):
        super().__init__()
        self.layers = nn.ModuleList([ConvLayer(layers[i], layers[i + 1])
            for i in range(len(layers) - 1)])
        self.out = nn.Linear(layers[-1], c)

    def forward(self, x):
        for l in self.layers: x = l(x) # ConvNet:  for l in self.layers: x = F.relu(l(x))
        x = F.adaptive_max_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        return F.log_softmax(self.out(x), dim=-1)
```

## BatchNorm

```python
# BatchNorm makes it really easy to train deeper networks.
# The last model, when we tried to add more layers, we had trouble training. 
# The reason we had trouble training was that if we used larger learning rates, it would go off to NaN and if we used smaller learning rate,
# it would take forever and doesn’t have a chance to explore properly — so it was not resilient.

# We can simply use nn.BatchNorm but to learn about it, we will write it from scratch.
class BnLayer(nn.Module):
    def __init__(self, ni, nf, stride=2, kernel_size=3):
        super().__init__()
        self.conv = nn.Conv2d(ni, nf, kernel_size=kernel_size, stride=stride,
                              bias=False, padding=1)
        self.a = nn.Parameter(torch.zeros(nf,1,1)) # all elements are 0
        self.m = nn.Parameter(torch.ones(nf,1,1)) # all elements are 1
        
    def forward(self, x): # (batch size, num of input filters, height, width)
        x = F.relu(self.conv(x))
        x_chan = x.transpose(0,1).contiguous().view(x.size(1), -1) # (num of output filters, height * width)
        if self.training:
            self.means = x_chan.mean(1)[:,None,None] # (num of output filters, 1, 1)
            self.stds  = x_chan.std (1)[:,None,None] # (num of output filters, 1, 1)
        return (x-self.means) / self.stds *self.m + self.a
```

```python
class ConvBnNet(nn.Module):
    def __init__(self, layers, c):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 10, kernel_size=5, stride=1, padding=2)
        self.layers = nn.ModuleList([BnLayer(layers[i], layers[i + 1])
            for i in range(len(layers) - 1)])
        self.out = nn.Linear(layers[-1], c)
        
    def forward(self, x):
        x = self.conv1(x)
        for l in self.layers: x = l(x)
        x = F.adaptive_max_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        return F.log_softmax(self.out(x), dim=-1)
```

## Deep BatchNorm

```python
class ConvBnNet2(nn.Module):
    def __init__(self, layers, c):
        super().__init__()
        # a bigger kernel size(5x5) and a stride of 1. The basic idea is that we want the first layer to have a richer input.
        self.conv1 = nn.Conv2d(3, 10, kernel_size=5, stride=1, padding=2) # Since padding = kernel_size — 1 / 2 and stride=1, the input size is the same as the output size
        self.layers = nn.ModuleList([BnLayer(layers[i], layers[i+1]) # stride=2, kernel_size=3, padding=1
            for i in range(len(layers) - 1)])
        self.layers2 = nn.ModuleList([BnLayer(layers[i+1], layers[i + 1], 1) # output size is same as input size since stride=1, kernel_size=3, padding=1
            for i in range(len(layers) - 1)])
        self.out = nn.Linear(layers[-1], c)
        
    def forward(self, x):
        x = self.conv1(x)
        for l,l2 in zip(self.layers, self.layers2):
            x = l(x)
            x = l2(x)
        x = F.adaptive_max_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        return F.log_softmax(self.out(x), dim=-1)
```

## Resnet

```python
class ResnetLayer(BnLayer):
    # x.size() and super().forward(x).size() returns same size
    #     ex. Conv2d(20, 20, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1)) doesn't change size
    def forward(self, x): return x + super().forward(x) # super().forward(x) outputs residual
```

```python
# add bunch of layers and make it 3 times deeper, and it still trains beautifully just because of x + super().forward(x)
class Resnet(nn.Module):
    def __init__(self, layers, c):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 10, kernel_size=5, stride=1, padding=2)
        self.layers = nn.ModuleList([BnLayer(layers[i], layers[i+1]) #  standard convolution with stride=2, this is called a “bottleneck layer
            for i in range(len(layers) - 1)])
        self.layers2 = nn.ModuleList([ResnetLayer(layers[i+1], layers[i + 1], 1) # 1: stride=1
            for i in range(len(layers) - 1)])
        self.layers3 = nn.ModuleList([ResnetLayer(layers[i+1], layers[i + 1], 1) # 1: stride=1
            for i in range(len(layers) - 1)])
        self.out = nn.Linear(layers[-1], c)
        
    def forward(self, x):
        x = self.conv1(x)
        for l,l2,l3 in zip(self.layers, self.layers2, self.layers3):
            x = l3(l2(l(x)))
        x = F.adaptive_max_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        return F.log_softmax(self.out(x), dim=-1)
```

## Resnet 2

```python
class Resnet2(nn.Module):
    def __init__(self, layers, c, p=0.5):
        super().__init__()
        self.conv1 = BnLayer(3, 16, stride=1, kernel_size=7)
        self.layers = nn.ModuleList([BnLayer(layers[i], layers[i+1])
            for i in range(len(layers) - 1)])
        self.layers2 = nn.ModuleList([ResnetLayer(layers[i+1], layers[i + 1], 1)
            for i in range(len(layers) - 1)])
        self.layers3 = nn.ModuleList([ResnetLayer(layers[i+1], layers[i + 1], 1)
            for i in range(len(layers) - 1)])
        self.out = nn.Linear(layers[-1], c)
        self.drop = nn.Dropout(p) # added dropout
        
    def forward(self, x):
        x = self.conv1(x)
        for l,l2,l3 in zip(self.layers, self.layers2, self.layers3):
            x = l3(l2(l(x)))
        x = F.adaptive_max_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        x = self.drop(x)
        return F.log_softmax(self.out(x), dim=-1)
```

```python
# increased the size of features and added dropout
learn = ConvLearner.from_model_data(Resnet2([16, 32, 64, 128, 256], 10, 0.2), data)
```
